```python
import pandas as pd
import altair as alt
from vega_datasets import data
```

```python
alt.renderers.enable('notebook')
```

```python
movies_data = data.movies()

movies_data.sample(5)
```

```python
movies_data.dropna(inplace=True)
```

* In this plot we can see that th eproduction budget and average budget

* Let's see the count of films which is according to the the IMDB Rating. From this we can tell that, more number of films are with hight rating
* Darken the colour is with count

```python
alt.Chart(movies_data, height=400, width=600) \
   .mark_bar() \
   .encode(alt.X('IMDB_Rating', bin=True), 
           y='count()', 
           color='count()')
```

* Here we can see that the rotten tomatoes rating which is 70-80, is more. The color and length of the bar, both gives the same information

```python
alt.Chart(movies_data, height=400, width=600) \
   .mark_bar() \
   .encode(alt.X('Rotten_Tomatoes_Rating', bin=True), 
           y='count()', 
           color='count()')
```

* Finally if we will check the count of running time of the film and IMDB rating. Count of  Films which is running for 90-140 is more

```python
alt.Chart(movies_data, height=400, width=600) \
   .mark_bar() \
   .encode(alt.X('Running_Time_min', bin=True), 
           alt.Y('IMDB_Rating', bin=True), 
           color='count()')
```

* From here we can see that the relationship between running time of the film and IMDB rating. Basically it is showing some of the films have high IMDB rating but the running time is less. Some other films has less running time,  also has less IDB Rating.
* Basically the minimum running time to get good IMDB Rating is more than 90 min
* If the film is too long also peoples are giving less IMDB Rating
* Also whichever films has the high IMDB Rating then the average rotten_tomatoes rating also high

```python
alt.Chart(movies_data, height=400, width=600) \
   .mark_circle() \
   .encode(alt.X('Running_Time_min', bin=True), 
            alt.Y('IMDB_Rating', bin=True), 
            size='IMDB_Rating', 
            color='average(Rotten_Tomatoes_Rating)')
```

```python
alt.Chart(movies_data, height=400, width=600) \
   .mark_bar() \
   .encode(alt.X('Production_Budget', bin=True), 
           alt.Y('IMDB_Rating', bin=True), 
           color='count()')
```

```python
alt.Chart(movies_data, height=400, width=600) \
   .mark_square() \
   .encode(alt.X('Production_Budget', bin=True), 
           alt.Y('IMDB_Rating', bin=True), 
           color='count()')
```

```python
alt.Chart(movies_data, height=400, width=600) \
   .mark_rect() \
   .encode(alt.X('Production_Budget', bin=True), 
           alt.Y('Worldwide_Gross', bin=True), 
           color='average(Rotten_Tomatoes_Rating)')
```
