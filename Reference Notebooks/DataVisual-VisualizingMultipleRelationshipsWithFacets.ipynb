```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
automobile_data_processed = pd.read_csv('datasets/automobile_data_processed.csv')

automobile_data_processed.sample(5)
```

* We can see the price according to the number of fuel type

```python
plt.figure(figsize=(15,10))

fg = sns.FacetGrid(automobile_data_processed, 
                   col='fuel-type', 
                   height=7, aspect=1)

fg.map(plt.hist, 'price', color='red')

plt.show()
```


* Let's see here price and the horsepower, defining the fuel type as columns, and it differentiated with number of cylinders
* Also almost all the cars, the number of cylinder is four

```python
fg = sns.FacetGrid(automobile_data_processed, 
                   col='fuel-type', 
                   height=7, aspect=1)

fg.map(sns.scatterplot, 'highway-mpg', 'price', s=100)

fg.add_legend()
```

```python
fg = sns.FacetGrid(automobile_data_processed, 
                   col='fuel-type', 
                   hue='num-of-cylinders', 
                   height=7, aspect=1)

fg.map(sns.scatterplot, 'highway-mpg', 'price', s=100)

fg.add_legend()
```

```python
tips_data = pd.read_csv('datasets/tips.csv')

tips_data.sample(10)
```

```python
plt.figure(figsize=(15,10))

fg = sns.FacetGrid(tips_data, 
                   col='size', 
                   height=7, aspect=1)

fg.map(plt.hist, 'total_bill', color='green')

plt.show()
```

```python
fg = sns.FacetGrid(tips_data, 
                   row='time', 
                   height=6, aspect=1)

fg.map(sns.swarmplot, 'size', 'total_bill')

fg.add_legend()
```

```python
fg = sns.FacetGrid(tips_data, 
                   col='time', 
                   hue='sex',
                   height=6, aspect=1)

fg.map(sns.swarmplot, 'size', 'total_bill')

fg.add_legend()
```
