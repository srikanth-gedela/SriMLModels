#### Importing Libraries

```python
import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Importing Dataset

```python
tips_data = pd.read_csv('datasets/tips.csv')

tips_data.sample(10)
```

```python
tips_data.shape
```

#### Describe the data

```python
tips_data.describe()
```

```python
tips_data.groupby(by='time').mean()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='time', y='total_bill', data=tips_data)

plt.title('Bill and Tips')
plt.xticks(rotation=90)

plt.show()
```

* Finding mean of the supermarket rating

```python
mean = tips_data['tip'].mean()

mean
```

```python
tips_data['above average'] = (tips_data['tip'] - mean) > 0

tips_data[['tip', 'above average']].head(5)
```

```python
plt.figure(figsize=(12, 8))

sns.countplot('time', 
              hue='above average', 
              data=tips_data, 
              order = tips_data['time'].value_counts().index)

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot('time', 
              hue='sex', 
              data=tips_data, 
              order = tips_data['time'].value_counts().index)

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='sex', 
            y='total_bill', 
            data=tips_data, 
            palette='nipy_spectral')

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='size', 
            y='total_bill', 
            data=tips_data, 
            palette='nipy_spectral')

plt.title('Bill and Tips')

plt.show()
```

```python
sns.catplot(x='size', 
            y='total_bill', 
            hue='time', 
            data=tips_data, 
            kind='bar', height=6, aspect=2, 
            palette='CMRmap')

plt.title('Bill and Tips')

plt.show()
```

* Let's see about payment in branch, There is no much difference for the payment system on each branch

```python
plt.figure(figsize=(12, 8))

sns.stripplot(x='size',  
              y='tip',  
              data=tips_data)

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='size',  
              y='tip',  
              data=tips_data)

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='size',  
              y='total_bill', 
              hue='above average',
              data=tips_data)

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='size',  
              y='total_bill',  
              hue='above average',
              data=tips_data, dodge=True)

plt.title('Bill and Tips')

plt.show()
```

* Using this plot we can see that what kind of customers are there in different payment 

```python
plt.figure(figsize=(12, 8))

sns.violinplot(x='smoker', 
               y='total_bill', 
               hue='time', 
               data=tips_data, 
               palette='muted')

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.countplot('size', 
              data=tips_data, 
              order = tips_data['size'].value_counts().index )

plt.xticks(rotation=70)

plt.title('Bill and Tips')

plt.show()
```
